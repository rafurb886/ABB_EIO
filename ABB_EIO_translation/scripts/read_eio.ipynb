```python
columns_name = ['Name', 'SignalType', 'Device', 'Label', 'DeviceMap', 'Category', 'Access', 'Default', 'SafeLevel', 'EncType', 'MaxLog', 'MaxPhys', 'MaxPhysLimit', 'MaxBitVal', 'MinLog', 'MinPhys', 'MinPhysLimit', 'MinBitVal']
df = pd.DataFrame(data=None, columns=columns_name)
```

```python
test_roi = roi[0:3]
```

```python
with_apostrophe = ['Name', 'SignalType', 'Device', 'Label', 'DeviceMap', 'Category', 'Access', 'SafeLevel', 'EncType']
without_apostrophe = set(columns_name).difference(set(with_apostrophe))
without_apostrophe
```

```python
def find_params(string_line, columns_name):
    result_dict = dict()
    result_list = []
    for column in columns_name:
        if column in with_apostrophe:
            pattern = re.compile(f'-{column} "(.*?)"')
        elif column in without_apostrophe:
            pattern = re.compile(f'-{column} (\d*)')
        result = pattern.search(string_line)
        result_dict[column] = result.group(1) if result!=None else np.NaN
    return result_dict
    
```

```python
list_to_df=[]
for line_to_df in roi:
    to_df = find_params(line_to_df, columns_name)
    df = df.append(to_df, ignore_index=True)
df
```

```python
def regex_fnt(val):
    pattern = re.compile('(\d+).?')
    result = pattern.search(val)
    return result.group(1)
    
```

```python
df['DeviceMapToSort'] = df['DeviceMap'].apply(regex_fnt)
```

```python
df['DeviceMapToSort'] = df['DeviceMapToSort'].astype('int32')
```
